```python
# Import necessary Libaries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
df = pd.read_csv('Dataset/Solar_Power_Output.csv')

# Display first few rows of dataset
df.head(10)
```

```python
# Display total number of rows and columns
df.shape
```

```python
# Display Summary
df.describe()
```

```python
# Check the information of the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check duplicated values
df.duplicated().sum()
```

```python
df.columns
```

```python
# Plot distribution of power
plt.figure(figsize=(10,6))
plt.title('Distribution of generated power(KW)')
plt.xlabel('Generated Power(KW)')
plt.ylabel('Frequency')
sns.histplot(df['generated_power_kw'],bins=20, kde=True)
plt.show()
```

### Exploratory Data Analysis


```python
df[df.columns[:9]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
df[df.columns[9:18]].hist(bins=30, figsize=(15,10))
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
# Bivariate Analysis
# Scatter plot with target feature
plt.figure(figsize=(15,25))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power (kW)')
    plt.xlabel(column)
    plt.ylabel('Generated Power (kW)')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True,fmt='.2f')
```

```python
# Outliers
plt.figure(figsize=(15,25))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.show()
```
